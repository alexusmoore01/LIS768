```python
import csv
from collections import Counter
import matplotlib.pyplot as plt

# Read the CSV file using iso-8859-1 encoding
with open("/Users/alexusmoore/Desktop/IranOnlyEnglish.csv", 'r', encoding='iso-8859-1') as file:
    reader = csv.reader(file)
    # Concatenate all tweets into a single string
    tweets = ' '.join([tweet[0] for tweet in reader])

# Split the string into individual words and count their frequency
word_counts = Counter(tweets.split())

# Get the top 10 most common words and their frequency
top_words = word_counts.most_common(5)

# Create a bar chart of the top 10 most common words
words = [word for word, count in top_words]
counts = [count for word, count in top_words]

plt.bar(words, counts)
plt.title('Top 5 Most Used Hashtags')
plt.xlabel('Hashtags')
plt.ylabel('Count')

# Rotate the y-axis label 45 degrees
plt.setp(plt.gca().get_xticklabels(), rotation=45, ha='right')

plt.show()
```

```python
import csv
from collections import Counter
import matplotlib.pyplot as plt

# Read the CSV file using iso-8859-1 encoding
with open("/Users/alexusmoore/Desktop/IranOnlyEnglish.csv", 'r', encoding='iso-8859-1') as file:
    reader = csv.reader(file)
    # Concatenate all tweets into a single string
    tweets = ' '.join([tweet[0] for tweet in reader])

# Split the string into individual words and count their frequency
word_counts = Counter(tweets.split())

# Get the top 5 most common words and their frequency
top_words = word_counts.most_common(5)

# Create a pie chart of the top 5 most common words
words = [word for word, count in top_words]
counts = [count for word, count in top_words]

plt.pie(counts, labels=words, autopct='%1.1f%%')
plt.title('Top 5 Most Common Hashtags')
plt.show()
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Read in the data from a CSV file
df = pd.read_csv("/Users/alexusmoore/Desktop/IranOnlyEnglish.csv", 'r', encoding='iso-8859-1')

# Create a list to store the polarity values
polarity_list = []

# Iterate through each tweet and calculate its polarity score
for tweet in df["text"]:
    polarity = TextBlob(tweet).sentiment.polarity
    polarity_list.append(polarity)

# Categorize each polarity value into positive, negative or neutral
positive = 0
negative = 0
neutral = 0
for polarity in polarity_list:
    if polarity > 0:
        positive += 1
    elif polarity < 0:
        negative += 1
    else:
        neutral += 1

# Create a pie chart to show the distribution of sentiment
labels = ['Positive', 'Negative', 'Neutral']
sizes = [positive, negative, neutral]
colors = ['green', 'red', 'gray']
explode = (0.1, 0, 0)  # explode the 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title("Sentiment Analysis Results")
plt.show()
```
